```python
#Import Pandas
import pandas as pd
import seaborn as sns
```

```python
#Define the dataset in a dataframe
dataset = pd.read_csv("nyc-rolling-sales.csv")

#List headings so that we can copy them and change them easily
list(dataset.columns.values)
```

```python
#Rename the columns of the dataset
dataset.columns = ['INDEX',
 'BOROUGH',
 'NEIGHBORHOOD',
 'BUILDING_CLASS_CATEGORY',
 'TAX_CLASS_AT_PRESENT',
 'BLOCK',
 'LOT',
 'EASE-MENT',
 'BUILDING_CLASS_AT_PRESENT',
 'ADDRESS',
 'APARTMENT_NUMBER',
 'ZIP_CODE',
 'RESIDENTIAL_UNITS',
 'COMMERCIAL_UNITS',
 'TOTAL_UNITS',
 'LAND_SQUARE_FEET',
 'GROSS_SQUARE_FEET',
 'YEAR_BUILT',
 'TAX_CLASS_AT_TIME_OF_SALE',
 'BUILDING_CLASS_AT_TIME_OF_SALE',
 'SALE_PRICE',
 'SALE_DATE']
list(dataset.columns.values)
```

```python
#Check the data table
dataset.head (100)
```

```python
#Remove all rows of data that have blank values for either, "SALE PRICE", "LAND SQUARE FEET" or "GROSS SQUARE FEET"
dataset = dataset[dataset.SALE_PRICE != " -  "]
dataset = dataset[dataset.GROSS_SQUARE_FEET != " -  "]
dataset = dataset[dataset.LAND_SQUARE_FEET != " -  "]
dataset.head(100)
```

```python
#Change the format of the cell value into an integer for key data fields
dataset.SALE_PRICE = dataset.SALE_PRICE.astype(int)
dataset.LAND_SQUARE_FEET = dataset.LAND_SQUARE_FEET.astype(int)
dataset.GROSS_SQUARE_FEET = dataset.GROSS_SQUARE_FEET.astype(int)
dataset.info()
```
