```python
import pandas as pd
```

```python
dataset = pd.read_csv("nyc-rolling-sales.csv")
```

```python
dataset.columns = ['INDEX',
 'BOROUGH',
 'NEIGHBORHOOD',
 'BUILDING_CLASS_CATEGORY',
 'TAX_CLASS_AT_PRESENT',
 'BLOCK',
 'LOT',
 'EASE-MENT',
 'BUILDING_CLASS_AT_PRESENT',
 'ADDRESS',
 'APARTMENT_NUMBER',
 'ZIP_CODE',
 'RESIDENTIAL_UNITS',
 'COMMERCIAL_UNITS',
 'TOTAL_UNITS',
 'LAND_SQUARE_FEET',
 'GROSS_SQUARE_FEET',
 'YEAR_BUILT',
 'TAX_CLASS_AT_TIME_OF_SALE',
 'BUILDING_CLASS_AT_TIME_OF_SALE',
 'SALE_PRICE',
 'SALE_DATE']
```

```python
dataset = dataset[dataset.SALE_PRICE != " -  "]
dataset = dataset[dataset.GROSS_SQUARE_FEET != " -  "]
dataset = dataset[dataset.LAND_SQUARE_FEET != " -  "]
```

```python
dataset.SALE_PRICE = dataset.SALE_PRICE.astype(int)
dataset.LAND_SQUARE_FEET = dataset.LAND_SQUARE_FEET.astype(int)
dataset.GROSS_SQUARE_FEET = dataset.GROSS_SQUARE_FEET.astype(int)
dataset.info()
```
