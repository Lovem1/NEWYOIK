```python
#Define the dataset in a dataframe
dataset = pd.read_csv("nyc-rolling-sales.csv")

#List headings so that we can copy them and change them easily
list(dataset.columns.values)
```

```python
#Rename the columns of the dataset
dataset.columns = ['INDEX',
 'BOROUGH',
 'NEIGHBORHOOD',
 'BUILDING_CLASS_CATEGORY',
 'TAX_CLASS_AT_PRESENT',
 'BLOCK',
 'LOT',
 'EASE-MENT',
 'BUILDING_CLASS_AT_PRESENT',
 'ADDRESS',
 'APARTMENT_NUMBER',
 'ZIP_CODE',
 'RESIDENTIAL_UNITS',
 'COMMERCIAL_UNITS',
 'TOTAL_UNITS',
 'LAND_SQUARE_FEET',
 'GROSS_SQUARE_FEET',
 'YEAR_BUILT',
 'TAX_CLASS_AT_TIME_OF_SALE',
 'BUILDING_CLASS_AT_TIME_OF_SALE',
 'SALE_PRICE',
 'SALE_DATE']
list(dataset.columns.values)
```

```python
#Create a column that calculates price per square foot for each property
df['Price_per_sqft'] = df['SALE_PRICE']/df['GROSS_SQUARE_FEET']
```

```python
df.head(5)
```

```python
#Calculate the average price per square foot, by Neighbourhood. Make this a dataset.
avgprice = df.groupby("NEIGHBORHOOD").mean()
```

```python
avgprice.head(5)
```

```python
def plot_bar_x():
    plt.bar(avgprice.index, avgprice.SALE_PRICE)
    plt.xlabel('HOOD', fontsize=5)
    plt.ylabel('Avg. Price per Sqft.', fontsize=5)
    plt.xticks(avgprice.index, avgprice.SALE_PRICE, fontsize=5, rotation=30)
    plt.title('Average price per square foot of New York properties')
    plt.show()
```

```python
plot_bar_x()
```
